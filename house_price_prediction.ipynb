```
import turicreate
```

```
sales = turicreate.SFrame('home_data.csv')
```

```
sales
```

## **Exploring the data for housing sales**

```
sales.show()
```

```
turicreate.visualization.scatter(sales['sqft_living'],sales['price'],xlabel="square feet living",ylabel="price")
```

##**Create a simple regression model**

```
train_data,test_data=sales.random_split(.8,seed=0)
```

##**Build the regression model**

```
sqft_model=turicreate.linear_regression.create(train_data,target="price",features=['sqft_living'])
```

##**Evaluate the simple model**

```
print(test_data["price"].mean())
```

```
sqft_model.evaluate(test_data)
```

##**Prediction Visualization**

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.plot(test_data['sqft_living'],test_data["price"],".",test_data["sqft_living"],sqft_model.predict(test_data))
```

```
sqft_model.coefficients
```
